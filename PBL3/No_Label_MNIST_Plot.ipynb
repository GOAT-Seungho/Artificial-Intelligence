```python
import os
import struct
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_no_label(images):
    with open(images, 'rb') as imgpath:
        magic, num, rows, cols = struct.unpack(">IIII", imgpath.read(16))
        images = np.fromfile(imgpath, dtype=np.uint8).reshape(60000, 28, 28)
    return images
```

```python
X = read_no_label(os.getcwd()+'/../data/testall-images-idx3-ubyte')
```

```python
print(X.shape)
```

```python
plt.figure(figsize=(20,4))
for index, image in enumerate(X[0:10]):
    plt.subplot(1, 10, index + 1)
    plt.imshow(image, cmap=plt.cm.gray)
```
